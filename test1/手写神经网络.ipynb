- **identify_function**: æ’ç­‰å‡½æ•°ï¼Œç›´æ¥è¿”å›è¾“å…¥å€¼ã€‚
- **sigmoid**: Sigmoidæ¿€æ´»å‡½æ•°ï¼Œç”¨äºå°†è¾“å…¥å€¼å‹ç¼©åˆ°0å’Œ1ä¹‹é—´ï¼Œå¸¸ç”¨äºç¥ç»ç½‘ç»œçš„éšè—å±‚ã€‚
- **softmax**: Softmaxå‡½æ•°ï¼Œç”¨äºè®¡ç®—å¤šåˆ†ç±»é—®é¢˜ä¸­æ¯ä¸ªç±»åˆ«çš„æ¦‚ç‡ï¼Œé€šå¸¸ç”¨äºç¥ç»ç½‘ç»œçš„è¾“å‡ºå±‚ã€‚

### æŸå¤±å‡½æ•°
- **cross_entropy_error**: äº¤å‰ç†µæŸå¤±å‡½æ•°ï¼Œç”¨äºè®¡ç®—é¢„æµ‹å€¼å’ŒçœŸå®å€¼ä¹‹é—´çš„è¯¯å·®ï¼Œå¸¸ç”¨äºåˆ†ç±»é—®é¢˜çš„æŸå¤±è®¡ç®—ã€‚

### åˆå§‹åŒ–ç½‘ç»œ
- **init_network**: åˆå§‹åŒ–ç¥ç»ç½‘ç»œçš„æƒé‡å’Œåç½®ï¼Œè¿”å›ä¸€ä¸ªåŒ…å«è¿™äº›å‚æ•°çš„å­—å…¸ã€‚è¿™é‡Œä½¿ç”¨çš„æ˜¯ç»™å®šçš„å›ºå®šå‚æ•°ã€‚

### å‰å‘ä¼ æ’­
- **forward**: å‰å‘ä¼ æ’­å‡½æ•°ï¼Œä¾æ¬¡è®¡ç®—æ¯ä¸€å±‚çš„åŠ æƒå’ŒåŠæ¿€æ´»å‡½æ•°ï¼Œæœ€ç»ˆè¾“å‡ºç»è¿‡softmaxå¤„ç†çš„ç»“æœï¼ŒåŒæ—¶è¿”å›éšè—å±‚çš„è¾“å‡ºã€‚

### åå‘ä¼ æ’­
- **backward**: åå‘ä¼ æ’­å‡½æ•°ï¼Œè®¡ç®—æŸå¤±å‡½æ•°ç›¸å¯¹äºæ¯ä¸ªå‚æ•°çš„æ¢¯åº¦ï¼Œå¹¶ä½¿ç”¨æ¢¯åº¦ä¸‹é™æ³•æ›´æ–°ç½‘ç»œçš„æƒé‡å’Œåç½®ã€‚

### é¢„æµ‹
- **predict**: é¢„æµ‹å‡½æ•°ï¼Œä½¿ç”¨è®­ç»ƒåçš„ç½‘ç»œè¿›è¡Œé¢„æµ‹ï¼Œè¿”å›ç½‘ç»œçš„è¾“å‡ºç»“æœã€‚

### è®­ç»ƒè¿‡ç¨‹
- **è®­ç»ƒè¿‡ç¨‹**: åœ¨è®­ç»ƒè¿‡ç¨‹ä¸­ï¼Œè¾“å…¥æ•°æ®å’ŒçœŸå®æ ‡ç­¾é€šè¿‡å‰å‘ä¼ æ’­è®¡ç®—è¾“å‡ºï¼Œç„¶åé€šè¿‡åå‘ä¼ æ’­è®¡ç®—æ¢¯åº¦å¹¶æ›´æ–°ç½‘ç»œçš„å‚æ•°ã€‚è®­ç»ƒå¤šä¸ªå‘¨æœŸï¼ˆepochsï¼‰ï¼Œä¸æ–­ä¼˜åŒ–ç½‘ç»œå‚æ•°ä»¥å‡å°‘æŸå¤±å‡½æ•°å€¼ã€‚

### é¢„æµ‹è¿‡ç¨‹
- **é¢„æµ‹è¿‡ç¨‹**: ä½¿ç”¨è®­ç»ƒåçš„ç½‘ç»œå¯¹æ–°è¾“å…¥æ•°æ®è¿›è¡Œé¢„æµ‹ï¼Œè¾“å‡ºåˆ†ç±»ç»“æœçš„æ¦‚ç‡åˆ†å¸ƒã€‚

```python
import numpy as np

def identify_function(x):
    """æ’ç­‰å‡½æ•°ï¼Œç›´æ¥è¿”å›è¾“å…¥å€¼"""
    return x

def sigmoid(x):
    """Sigmoidæ¿€æ´»å‡½æ•°"""
    return 1 / (1 + np.exp(-x))

def softmax(a):
    """Softmaxå‡½æ•°"""
    c = np.max(a)  # é˜²æ­¢æº¢å‡º
    exp_a = np.exp(a - c)  # å‡å»æœ€å¤§å€¼c
    sum_exp_a = np.sum(exp_a)  # æŒ‡æ•°å‡½æ•°çš„å’Œ
    y = exp_a / sum_exp_a  # å½’ä¸€åŒ–
    return y

def init_network():
    """åˆå§‹åŒ–ç¥ç»ç½‘ç»œï¼Œè¿”å›ä¸€ä¸ªåŒ…å«æƒé‡å’Œåç½®çš„å­—å…¸"""
    network = {}
    network['W1'] = np.array([[0.1, 0.3, 0.5], [0.2, 0.4, 0.6]])
    network['b1'] = np.array([0.1, 0.2, 0.3])
    network['W2'] = np.array([[0.1, 0.4], [0.2, 0.5], [0.3, 0.6]])
    network['b2'] = np.array([0.1, 0.2])
    network['W3'] = np.array([[0.1, 0.3], [0.2, 0.4]])
    network['b3'] = np.array([0.1, 0.2])
    return network

def forward(network, x):
    """å‰å‘ä¼ æ’­å‡½æ•°ï¼Œè®¡ç®—å¹¶è¿”å›è¾“å‡º"""
    W1, W2, W3 = network['W1'], network['W2'], network['W3']
    b1, b2, b3 = network['b1'], network['b2'], network['b3']
    
    # ç¬¬ä¸€å±‚
    a1 = np.dot(x, W1) + b1
    z1 = sigmoid(a1)
    
    # ç¬¬äºŒå±‚
    a2 = np.dot(z1, W2) + b2
    z2 = sigmoid(a2)
    
    # ç¬¬ä¸‰å±‚
    a3 = np.dot(z2, W3) + b3
    y = softmax(a3)
    
    return y

# åˆå§‹åŒ–ç½‘ç»œ
network = init_network()

# è¾“å…¥æ•°æ®
x = np.array([1.0, 0.5])

# å‰å‘ä¼ æ’­ï¼Œè®¡ç®—è¾“å‡º
y = forward(network, x)

# æ‰“å°è¾“å‡º
print(y)
```

- **sigmoid**: Sigmoidæ¿€æ´»å‡½æ•°ï¼Œç”¨äºå°†è¾“å…¥å€¼å‹ç¼©åˆ°0å’Œ1ä¹‹é—´ï¼Œå¸¸ç”¨äºç¥ç»ç½‘ç»œçš„éšè—å±‚ã€‚
- **softmax**: Softmaxå‡½æ•°ï¼Œç”¨äºè®¡ç®—å¤šåˆ†ç±»é—®é¢˜ä¸­æ¯ä¸ªç±»åˆ«çš„æ¦‚ç‡ï¼Œé€šå¸¸ç”¨äºç¥ç»ç½‘ç»œçš„è¾“å‡ºå±‚ã€‚

### æŸå¤±å‡½æ•°

### åˆå§‹åŒ–ç½‘ç»œ
- **init_network**: åˆå§‹åŒ–ç¥ç»ç½‘ç»œçš„æƒé‡å’Œåç½®ï¼Œè¿”å›ä¸€ä¸ªåŒ…å«è¿™äº›å‚æ•°çš„å­—å…¸ã€‚ä¸ä¹‹å‰çš„ç‰ˆæœ¬ä¸åŒï¼Œè¿™é‡Œä½¿ç”¨éšæœºæ•°è¿›è¡Œåˆå§‹åŒ–ï¼Œè€Œä¸æ˜¯å›ºå®šå‚æ•°ã€‚

### å‰å‘ä¼ æ’­
- **forward**: å‰å‘ä¼ æ’­å‡½æ•°ï¼Œä¾æ¬¡è®¡ç®—æ¯ä¸€å±‚çš„åŠ æƒå’ŒåŠæ¿€æ´»å‡½æ•°ï¼Œæœ€ç»ˆè¾“å‡ºç»è¿‡softmaxå¤„ç†çš„ç»“æœï¼ŒåŒæ—¶è¿”å›éšè—å±‚çš„è¾“å‡ºã€‚

### åå‘ä¼ æ’­
- **backward**: åå‘ä¼ æ’­å‡½æ•°ï¼Œè®¡ç®—æŸå¤±å‡½æ•°ç›¸å¯¹äºæ¯ä¸ªå‚æ•°çš„æ¢¯åº¦ï¼Œå¹¶ä½¿ç”¨æ¢¯åº¦ä¸‹é™æ³•æ›´æ–°ç½‘ç»œçš„æƒé‡å’Œåç½®ã€‚

### é¢„æµ‹
- **predict**: é¢„æµ‹å‡½æ•°ï¼Œä½¿ç”¨è®­ç»ƒåçš„ç½‘ç»œè¿›è¡Œé¢„æµ‹ï¼Œè¿”å›ç½‘ç»œçš„è¾“å‡ºç»“æœã€‚

### è®­ç»ƒè¿‡ç¨‹
- **è®­ç»ƒè¿‡ç¨‹**: åœ¨è®­ç»ƒè¿‡ç¨‹ä¸­ï¼Œè¾“å…¥æ•°æ®å’ŒçœŸå®æ ‡ç­¾é€šè¿‡å‰å‘ä¼ æ’­è®¡ç®—è¾“å‡ºï¼Œç„¶åé€šè¿‡åå‘ä¼ æ’­è®¡ç®—æ¢¯åº¦å¹¶æ›´æ–°ç½‘ç»œçš„å‚æ•°ã€‚è®­ç»ƒå¤šä¸ªå‘¨æœŸï¼ˆepochsï¼‰ï¼Œä¸æ–­ä¼˜åŒ–ç½‘ç»œå‚æ•°ä»¥å‡å°‘æŸå¤±å‡½æ•°å€¼ã€‚

### é¢„æµ‹è¿‡ç¨‹
- **é¢„æµ‹è¿‡ç¨‹**: ä½¿ç”¨è®­ç»ƒåçš„ç½‘ç»œå¯¹æ–°è¾“å…¥æ•°æ®è¿›è¡Œé¢„æµ‹ï¼Œè¾“å‡ºåˆ†ç±»ç»“æœçš„æ¦‚ç‡åˆ†å¸ƒã€‚

### åŒºåˆ«æ€»ç»“
1. **åˆå§‹åŒ–ç½‘ç»œ**: åŸå§‹ç‰ˆæœ¬ä½¿ç”¨å›ºå®šå‚æ•°è¿›è¡Œåˆå§‹åŒ–ï¼Œä¿®æ”¹ç‰ˆæœ¬ä½¿ç”¨éšæœºæ•°è¿›è¡Œåˆå§‹åŒ–ã€‚
2. **åå‘ä¼ æ’­**: ä¸¤ä¸ªç‰ˆæœ¬éƒ½å®ç°äº†åå‘ä¼ æ’­å’Œå‚æ•°æ›´æ–°ï¼Œä½†ç”±äºåˆå§‹åŒ–å‚æ•°çš„ä¸åŒï¼Œè®­ç»ƒçš„å…·ä½“è¿‡ç¨‹å’Œç»“æœä¼šæœ‰æ‰€ä¸åŒã€‚
3. **æ•´ä½“æ¡†æ¶**: å…¶ä½™éƒ¨åˆ†åŸºæœ¬ä¸€è‡´ï¼Œéƒ½åŒ…æ‹¬å‰å‘ä¼ æ’­ã€åå‘ä¼ æ’­ã€è®­ç»ƒå’Œé¢„æµ‹åŠŸèƒ½ã€‚

```python
import numpy as np

def identify_function(x):
    """æ’ç­‰å‡½æ•°ï¼Œç›´æ¥è¿”å›è¾“å…¥å€¼"""
    return x

def sigmoid(x):
    """Sigmoidæ¿€æ´»å‡½æ•°"""
    return 1 / (1 + np.exp(-x))

def softmax(a):
    """Softmaxå‡½æ•°"""
    c = np.max(a)  # é˜²æ­¢æº¢å‡º
    exp_a = np.exp(a - c)  # å‡å»æœ€å¤§å€¼c
    sum_exp_a = np.sum(exp_a)  # æŒ‡æ•°å‡½æ•°çš„å’Œ
    y = exp_a / sum_exp_a  # å½’ä¸€åŒ–
    return y

def cross_entropy_error(y, t):
    """äº¤å‰ç†µæŸå¤±å‡½æ•°"""
    if y.ndim == 1:
        t = t.reshape(1, t.size)
        y = y.reshape(1, y.size)
    batch_size = y.shape[0]
    return -np.sum(t * np.log(y + 1e-7)) / batch_size

def init_network():
    """åˆå§‹åŒ–ç¥ç»ç½‘ç»œï¼Œè¿”å›ä¸€ä¸ªåŒ…å«æƒé‡å’Œåç½®çš„å­—å…¸"""
    network = {}
    # éšæœºåˆå§‹åŒ–æƒé‡å’Œåç½®ï¼Œæ‰“ç ´å¯¹ç§°æ€§
    network['W1'] = np.random.randn(2, 3)
    network['b1'] = np.random.randn(3)
    network['W2'] = np.random.randn(3, 2)
    network['b2'] = np.random.randn(2)
    network['W3'] = np.random.randn(2, 2)
    network['b3'] = np.random.randn(2)
    return network

def forward(network, x):
    """å‰å‘ä¼ æ’­å‡½æ•°ï¼Œè®¡ç®—å¹¶è¿”å›è¾“å‡º"""
    W1, W2, W3 = network['W1'], network['W2'], network['W3']
    b1, b2, b3 = network['b1'], network['b2'], network['b3']
    
    # ç¬¬ä¸€å±‚
    a1 = np.dot(x, W1) + b1
    z1 = sigmoid(a1)
    
    # ç¬¬äºŒå±‚
    a2 = np.dot(z1, W2) + b2
    z2 = sigmoid(a2)
    
    # ç¬¬ä¸‰å±‚
    a3 = np.dot(z2, W3) + b3
    y = softmax(a3)
    
    return y, z1, z2

def backward(network, x, t, y, z1, z2, learning_rate=0.1):
    """åå‘ä¼ æ’­å¹¶æ›´æ–°å‚æ•°"""
    W1, W2, W3 = network['W1'], network['W2'], network['W3']
    b1, b2, b3 = network['b1'], network['b2'], network['b3']
    
    # è¾“å‡ºå±‚æ¢¯åº¦
    dy = (y - t) / y.shape[0]
    
    # ç¬¬äºŒå±‚åˆ°è¾“å‡ºå±‚æ¢¯åº¦
    dW3 = np.dot(z2.T, dy)
    db3 = np.sum(dy, axis=0)
    
    # ç¬¬ä¸€å±‚åˆ°ç¬¬äºŒå±‚æ¢¯åº¦
    dz2 = np.dot(dy, W3.T)
    da2 = dz2 * z2 * (1 - z2)
    dW2 = np.dot(z1.T, da2)
    db2 = np.sum(da2, axis=0)
    
    # è¾“å…¥å±‚åˆ°ç¬¬ä¸€å±‚æ¢¯åº¦
    dz1 = np.dot(da2, W2.T)
    da1 = dz1 * z1 * (1 - z1)
    dW1 = np.dot(x.T, da1)
    db1 = np.sum(da1, axis=0)
    
    # å‚æ•°æ›´æ–°
    W1 -= learning_rate * dW1
    b1 -= learning_rate * db1
    W2 -= learning_rate * dW2
    b2 -= learning_rate * db2
    W3 -= learning_rate * dW3
    b3 -= learning_rate * db3
    
    # æ›´æ–°ç½‘ç»œå‚æ•°
    network['W1'], network['b1'] = W1, b1
    network['W2'], network['b2'] = W2, b2
    network['W3'], network['b3'] = W3, b3

def predict(network, x):
    """é¢„æµ‹å‡½æ•°"""
    y, _, _ = forward(network, x)
    return y
```

```python
# å°è¯•è®­ç»ƒå¹¶é¢„æµ‹
# åˆå§‹åŒ–ç½‘ç»œ
network = init_network()

# è¾“å…¥æ•°æ®
x = np.array([[1.0, 0.5]])
t = np.array([[0, 1]])  # çœŸå®æ ‡ç­¾ï¼Œä½¿ç”¨one-hotç¼–ç 

# è®­ç»ƒè¿‡ç¨‹
epochs = 1000
for epoch in range(epochs):
    y, z1, z2 = forward(network, x)
    loss = cross_entropy_error(y, t)
    backward(network, x, t, y, z1, z2)
    if epoch % 100 == 0:
        print(f"Epoch {epoch}, Loss: {loss}")

# é¢„æµ‹
x_test = np.array([[1.0, 0.5]])
y_test = predict(network, x_test)
print("Predicted:", y_test)
```
